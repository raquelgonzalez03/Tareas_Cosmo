```python
import camb
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interpolate
from scipy.integrate import quad
```

```python
#definicion de la cosmologia 
H0 = 67.4 
omega_m = 0.315
omega_ch2 = 0.120
omega_bh2 = 0.024
ns = 0.965
tau = 0.054
```

```python
pars = camb.CAMBparams()
pars.set_cosmology(H0 = H0, ombh2 = omega_bh2, omch2= omega_ch2, tau = tau)
```

```python
pars.set_dark_energy(w=-1)
```

```python
results = camb.get_results(pars)
results
```

```python
#ejercicio 1: calcular el parametro de Hubble 
#esto lo hacemos usando una funcion especifica de CAMB 
#definimos el redshift
zmax = 3000.0
Nz = 6000
z = np.linspace(0.0, zmax, Nz)
a = 1.0 / (1.0 + z)
```

```python
Hz = results.hubble_parameter(z)
Hz
```

```python
#ejercicio 2: calcular el a(t) se hizo de forma analitica pero ahora lo vamos a comprobar numericamente
Mpc_to_km= 3.085677581491367e19
H_si = Hz / Mpc_to_km


ord_idx= np.argsort(a)
a_sorted = a [ord_idx]
H_sorted = H_si[ord_idx]


H_func = interpolate(a_sorted, H_sorted, kind = 'cubic', bounds_error=False, fill_value='extrapolate')
```
